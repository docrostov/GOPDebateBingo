```python
# Basic bingo code is cribbed lovingly from 
#   https://gist.github.com/matthiaseisen/2af0d71870427c6bdfd6
# 
# Edits include a shift to fill the free space with text 
#   & fill the board with items from "bingoStmts"

# Define my 5x5 matrix and sample the slices

def list_of_columns(
    numbers=range(1, len(bingoStmts)),
    num_of_columns=5,
    num_of_rows=5
):
    slice_length = len(numbers) // num_of_columns
    return [
        sorted(
            random.sample(
                numbers[i * slice_length: (i + 1) * slice_length],
                num_of_rows
            )
        )
        for i in range(num_of_columns)
    ]

# Transpose the columns to build out my matrix

def list_of_rows(list_of_columns):
    return [list(row) for row in zip(*list_of_columns)]

# Insert the free spaces and the GOP bingo statements

def insert_free_spaces(numbers, coords=[(2, 2)]):
    return [
        [
            bingoStmts[n] if not (x, y) in coords else "AN ANSWER GOES LONG!"
            for x, n in enumerate(numbers[y])
        ]
        for y in range(len(numbers))
    ]

def prepend_title_row(numbers):
    return [['B', 'I', 'N', 'G', 'O']] + numbers

def card_data():
    return prepend_title_row(
        insert_free_spaces(
            list_of_rows(
                list_of_columns()
            )
        )
    )
```

```python
class ListTable(list):
    """ Overridden list class which takes a 2-dimensional list of 
        the form [[1,2,3],[4,5,6]], and renders an HTML Table in 
        IPython Notebook. """
    
    def _repr_html_(self):
        html = ["<table style='border: 10px solid black; border-top: 0px solid black; border-collapse: collapse; font-family:verdana; font-size:75%; text-align: center; width:75%; padding:10px;'>"]
        i = 1
        for row in self:
            if i>1:
                html.append("<tr>")
            
                for col in row:
                    html.append("<td style='border: 2px solid #f1f1f1; text-align:center; width: 20%;'>{0}</td>".format(col))
            
                html.append("</tr>")
            else:
                html.append("<tr style='color: #ffffff; background-color: #000000';>")
                
                for col in row:
                    html.append("<td style='font-size: 200%; text-align:center; padding: 25px; width:20%;'>{0}</td>".format(col))
                
                html.append("</tr>")
                i+=1
        html.append("</table>")
        return ''.join(html)
    
ListTable(card_data())
```
